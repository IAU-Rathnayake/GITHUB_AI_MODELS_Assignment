### Task 1 : Sentiment Analysis with OpenAI GPT-4
Goal: Use the OpenAI GPT-4 model to analyze the sentiment of a given text (positive, negative, or neutral).

### Initailizing Imports


```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

### Setup

```python
load_dotenv()
token = os.environ.get("GITHUB_TOKEN")
endpoint = "https://models.github.ai/inference"
model_name = "mistral-ai/mistral-small-2503" 
```

```python
client = OpenAI(
    base_url=endpoint,
    api_key=token,
)
```

```python
def generate_code(user_prompt):
    
    system_prompt = "You are an expert Python programmer. Return only the python code requested. Do not add markdown backticks or explanations."

    response = client.chat.completions.create(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        model=model_name,
        temperature=0.1 # Low temperature for precise code
    )

    return response.choices[0].message.content
```

```python
prompt = "Write a Python function to reverse a string."
generated_code = generate_code(prompt)
```

```python
print(f"--- Task 02: Code Generation ({model_name}) ---\n")
print(f"Request: {prompt}")
print("Output: \n")
print(generated_code)
```
