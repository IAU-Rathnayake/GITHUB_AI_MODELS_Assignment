### Task 02: Code Generation with Anthropic Claude-2
Goal: Use Anthropic Claude-2 to generate a Python function that reverses a string.

### Initailizing Imports


```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

### Setup

```python
load_dotenv()
token = os.environ.get("GITHUB_TOKEN")
endpoint = "https://models.inference.ai.azure.com"
model_name = "gpt-4o"
```

```python
client = OpenAI(
    base_url=endpoint,
    api_key=token,
)
```

```python
def analyze_sentiment(text_input):
    
    system_prompt = "You are a sentiment analysis assistant. Analyze the input text and return ONLY one word: 'Positive', 'Negative', or 'Neutral'."

    response = client.chat.completions.create(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": text_input}
        ],
        model=model_name,
        temperature=0 
    )

    return response.choices[0].message.content



```

### Let's test the 3 scenarios 


```python
inputs = [
    "I'm Currently Learning AI and Machine Learning",           
    "This code is so confusing and broken.", 
    "The sky is blue today."                 
]

```

```python
print(f"--- Task 01: Sentiment Analysis ({model_name}) ---\n")

for text in inputs:
    result = analyze_sentiment(text)
    print(f"Input:  {text}")
    print(f"Result: {result}")
    print("-" * 30)
```
