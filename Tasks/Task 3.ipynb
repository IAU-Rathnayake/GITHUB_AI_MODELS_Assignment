### Task 03: Text Embedding with Mistral
Goal: Use a Mistral model to generate embeddings for a given text.

### Initailizing Imports


```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

### Setup

```python
load_dotenv()
token = os.environ.get("GITHUB_TOKEN")
endpoint = "https://models.github.ai/inference"
model_name = "text-embedding-3-small" #Mistral model aren't working
```

```python
client = OpenAI(
    base_url=endpoint,
    api_key=token,
)
```

```python
def get_string(user_prompt):
    
    response = client.embeddings.create(
        input = user_prompt,
        model=model_name,
    )
    return response.data[0].embedding
```

```python
prompt = "GitHub Models are powerful!"
vector =get_string(prompt)
print("Embedded Output: \n\n", vector[:5],"\n\nPrinting just the first 5 numbers to verify") 
```
